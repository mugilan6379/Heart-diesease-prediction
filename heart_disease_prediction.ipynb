```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import warnings
```

```python
warnings.filterwarnings('ignore')
hd_dataset = pd.read_csv('heart_disease_data.csv')
hd_dataset.head()
```

```python
hd_dataset.isnull().sum()
```

```python
correlation = hd_dataset.corr()
plt.figure(figsize=(10,10))
sns.heatmap(correlation,fmt='.1f',annot=True)
print(correlation['target'])
```

```python
X=hd_dataset.drop('target',axis=1)
Y=hd_dataset['target']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score,r2_score
from sklearn.ensemble import RandomForestRegressor,RandomForestClassifier


X_train, X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=4)
lr=RandomForestClassifier()
lr.fit(X_train,Y_train)
y_train_pred = lr.predict(X_train)
y_test_pred = lr.predict(X_test)

print('training: ',accuracy_score(Y_train,y_train_pred))
print('testing: ',accuracy_score(Y_test,y_test_pred))
```

```python
input_data = (59,1,0,140,177,0,1,162,1,0,2,1,3)

dat_arr  = np.asarray(input_data)
dat_arr_reshaped = dat_arr.reshape(1,-1)
prediction = lr.predict(dat_arr_reshaped)

if prediction[0] == 0:
    print('The person dont have heart disease')
else:
    print('The person have heart disease')
```
